

# Problem Statement

Atliquo, a new and ambitious bank, is planning to launch a credit card tailored to the growing market in India. The financial landscape in India is competitive, with established players like Axis Bank, ICICI Bank, and HDFC Bank dominating the market. To ensure the successful launch of this credit card, it is crucial to design a product that stands out and meets the needs of the target audience.

**Key Considerations:**
1. **Target Age Group:**
   - Should the card cater to students, working professionals, or seniors? Understanding the preferences and financial behavior of each group will help in tailoring the product effectively.

2. **Occupation:**
   - The type of job customers hold can influence their financial needs and spending habits. Features of the credit card should reflect these variations.

3. **Income Analysis:**
   - Customer income data is vital for determining appropriate cashback rates, promotions, and credit limits. These incentives must align with what is financially attractive and feasible for the target demographic.

4. **Payment Preferences:**
   - Analyzing the preferred payment methods across age groups will help identify the most relevant and convenient features for the target audience.

5. **Competitive Analysis:**
   - A thorough study of competitors' credit card offerings is essential to identify differentiators and gaps in the market. This will help in designing a product that provides unique value to customers.

---

**Objective:**
- Phase 1 : Identifying the ideal target market. The goal is to analyze the provided data to uncover insights and recommend strategies for Atliquo Bankâ€™s credit card laun
  ch. Customizing the credit card features based on customer needs and market trends.
- Phase 2 : After identifying the target market, we will run a trial on a small group and perform AB testing

To achieve this, the findings will be presented in a dashboard that is self-explanatory and designed for top-level management and the product strategy team. The dashboard will highlight key metrics, trends, and recommendations.

---

**Datasets:**
The analysis will utilize transactional data obtained from DATALIO, including the following datasets:

1. **customers.csv**
   - **cust_id:** Unique ID for each customer.
   - **name:** Name of the customer.
   - **gender:** Gender of the customer (Male, Female).
   - **age:** Age of the customer.
   - **location:** Location of the customer.
   - **occupation:** Occupation or profession of the customer.
   - **annual_income:** Annual income of the customer.
   - **marital_status:** Marital status (single, married).

2. **transactions.csv**
   - **trans_id:** Unique transaction ID.
   - **customer_id:** Customer ID linked to the customers table.
   - **tran_date:** Date of the transaction.
   - **tran_amount:** Total amount spent in the transaction.
   - **platform:** Platform used for the transaction.
   - **product_category:** Category of the product or service purchased.
   - **payment_type:** Type of payment used.

3. **credit_profiles.csv**
   - **cust_id:** Customer ID linked to the customers table.
   - **credit_score:** Credit score of the customer.
   - **credit_utilisation:** Percentage of the credit limit utilized by the customer.
   - **outstanding_debt:** Current debt of the customer.
   - **credit_inquiries_last_6_months:** Number of credit inquiries made in the last six months.
   - **credit_limit:** Total credit limit granted to the customer.

---

By analyzing these datasets, we aim to identify actionable insights and propose a strategic approach to the credit card launch, ensuring it meets market demands and positions Atliquo Bank as a competitive player in the Indian financial sector.

```python
# Set style for better visuals
sns.set(style="whitegrid")

# Define metrics for which we want to calculate the average per age_group
metrics = ['credit_score', 'credit_utilisation', 'credit_limit']

# Set up the 1x3 subplot grid
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle("Average Credit Metrics per Age Group", fontsize=16)

# Plot each metric in its respective subplot
for i, metric in enumerate(metrics):
    avg_metric = df_merge.groupby('age_group')[metric].mean().sort_values()
    sns.barplot(x=avg_metric.index, y=avg_metric.values, palette="viridis", ax=axes[i])
    
    axes[i].set_title(f'Average {metric.replace("_", " ").capitalize()} per Age Group')
    axes[i].set_xlabel('Age Group')
    axes[i].set_ylabel(f'Average {metric.replace("_", " ").capitalize()}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
